PAISES MAS CALIDOS

```python
import math
import pandas as pd
import matplotlib

paises_mundiales = pd.read_csv('countries of the world.csv',decimal = ',')

#Primeros 10 paises con mas temperatura del mundo
paises_mundiales.sort_values(by='Climate', ascending = False)[['Country','Climate']].head(10)

```

PAISES CON POBLACION MENOR A 3 MILLONES

```python
paises_menos_poblados = paises_mundiales['Population'] <= 3_000_000
paises_less = paises_mundiales[paises_menos_poblados]
#Ordenamos de mayor a menor
paises_less.sort_values(by='Population', ascending = False)
```

PAISES CON MAYOR LINEA COSTERA

```python
#Top ten de paises con mayor linea costera
paises_mundiales.sort_values(by='Coastline (coast/area ratio)', ascending = False).head(10)
```

PAISES CON MAS CANTIDAD DE CELULARES CADA 1000 PERSONAS

```python
#Top ten de paises con mas cantidad e telefonos
paises_mundiales.sort_values(by='Phones (per 1000)', ascending = False).head(10)
```

GRAFICO DE DATOS

```python

count = paises_mundiales.sort_values(by="Population", ascending = False).sort_values(by = "Service", ascending = False)[["Country","Climate","Agriculture","Service","Industry"]].head()
count.plot(x="Country", 
           y=["Industry","Agriculture"], 
           kind="bar",
          figsize = (16,10),
          color = ["blue","#191"])
```
