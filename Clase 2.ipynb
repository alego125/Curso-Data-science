Ver cual de los paises tiene 50 millones de personas.

Podemos filtrar el dataset, y esto lo podemos hacer a travez de algo llamado boolean indexing o indexado por buleano en español, esta es una de las herramientas mas poderosas que posee pandas, equivale a cuando hacemos un filtro en exel.

```python
countries_data.head(2)
```

```python
countries_data["Population"] < 50_000_000
```

Como podemos ver aca arriba tenemos que indicamos la columna a analizar, esta es population y al lado con un signo le inicamos que queremos todos aquellos que sean menores a 50 millones. Es decir lo que le estamos diciendo es que me analice cada fila de la columna population y me evalue cuales son mayores a 50 millones y en cada una de las filas me devolvera verdadero o falso segun sea que se cumple o no la condicion, todos estos datos nos lo devuelve en forma de lista

Como por ahi con numeros muy grandes puede complicarse sobre todo en la visualizacion de datos usamos guines bajos en las centenas para separar asi de esta manera es mas facil saber de que numeros se trata y python sigue identificando el numero como 50 millones.

Entonce ahora podemos guardar la serie en una variable y hacer un fltrado muy interesante:


```python
large_population = countries_data["Population"] > 50_000_000
large_population
```

```python
#ahora lo que haremos a continuacion es un indexado de booleanos, un slicing con large_population
countries_data[large_population]
```

Automaticamente como podemos ver aca arriba lo que sucede es que la lista se reduce drasticamente debido a que lo que hacemos es un ndexado de buleanos es decir al agregarle a la tabla una datalist de elementos booleanos lo que sucede es que se seleccionan solo aquellos que posean el valor true, por lo cual estos son los unicos valores que nos va a mostrar en la tabla.

Como podemos ver el valor correspondiente para el elemento en la posicion con indice 0 el cual es afganistan posee un valor de false en booleano por o cual no me lo va a mostrar. Asi ira analizando cada uno de los valores de cada uno de los indices para mostrarnos o no lo que cotiene segun de verdadero o falso.

Y esto que vimos aca arriba no es lo unico que podemos hacer, si no que podemos por ejemplo guardar la informacion en una variable copia y luego asi como tenemos coutries_data, tambien tener large_coutries, por ejemplo.


```python
large_countries = countries_data[large_population]
large_countries
```

Entonces con large_countries puedo hacer lo que quiera asi como teniamos a countries_data, esta tambien es una variable con datos que podemos trabajar de la misma manera, solaente que tendra nada mas los datos que sean con poblaciones mayores a 50 millones de personas.

```python
large_countries.head()
```

Asi vemos los primeros 5 elementos e esta nueva tabla de elementos, pero como podemos ver esta tabla no representa los prmeros 5 elementos con mas poblacion.

Entonce si quisiramos ordenar todo esto en orden de poblacion, para esto tendremos la funcion sort_values(), dos parametros importantes que tiene esta funcion son el by y el ascendig, (recordar que si no sabemos los parametros que trabaja una funcion solo basta con posicionarse sobre la funcion y presionar shift + tab para que se despliege el menu con los parametros de la funcion que podemos utilizar).

Si por ejemplo coloco:

```python
large_countries.sort_values(by = "Population")
#Asi por default va a ser ascendente por lo que nos va a mostrar en orden creciente los paises con su poblacion 
#Aca entra en juego otro parametro que es el ascendig que por default va a ser siempre verdadero y esto nos provoca lo dicho antes
```

Entonces si yo lo quiero hacer ascendente lo que debo hacer es pasar por parametro el siguiente parametro:

```python
large_countries.sort_values(by="Population", ascending = False)
#Asi de esta manera nos muestra los datos de manera descendente en cantidad de poblacion 
```

Despues a esto le podemos concatenar lo que queramos por ejemplo socar el top 5 de paises con mas poblacion del mundo, con el metodo head()

```python
large_countries.sort_values(by="Population", ascending = False).head()
```

Tambien podemos usar el doble criterio de ordenacion, por ejemplo podemos decir que ordene en relacion al la poblacion y a la migracion de la siguiente manera:

```python
large_countries.sort_values(by=["Population","Net migration"], ascending = False).head()
```

Otra cosa que podemos hacer es usar lo que hemos ordenado y usarlo para reordenar con respecto a otro parametro por ejemplo sevice, entonces ya quedaria ordenado por servicios del pais y como para desempatar se usaria la poblacion por ejemplo:

```python
large_countries.sort_values(by="Population", ascending = False).head().sort_values(by = "Service", ascending = False)
#Asi de esta manera nos muestra los paises ascendentemente en orde a sus sitemas de servicios
```

Algo que tambien se puede realizar es el guardado de todo esto en una varibale para despues ser utilizada, por ejemplo, tambien podemos realizar un indexado de columnas para solo nos muestre las ultimas tres columnas

```python
large_countries_services = large_countries.sort_values(by="Population", ascending = False).sort_values(by = "Service", ascending = False)[["Country","Climate","Agriculture","Service","Industry"]]
large_services_countries = large_countries_services.head()
large_services_countries
```

Asi de esta manera lo que hacemos es mostrar la informacion de las columnas que queremos de los elementos que queremos ordenados en el orden que queremos.

Entonces podemos ver que hemos sacado varios insites, ya sacamos los paises con mas poblacion, cual es el pais con mas infraestructura de servicio, entonces ya dejamos un dataset bien preparad para la siguiete parte que ya seriea el comenzar a plotear.
